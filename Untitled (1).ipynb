```python
#去掉没有薪资待遇的信息
origin_data = origin_data.dropna(subset = ['薪资待遇'],axis=0, how='any')
```

```python
#统计计算机岗位性质
def arithmatic(ele):
    if ("算法" in ele) or ("机器学习"in ele) or ("人工智能" in ele) or ("深度学习" in ele) or ("NLP" in ele)or ("视觉" in ele):
        return "算法岗"
    elif "后端" in ele:
        return '后端'
    elif '前端' in ele:
        return '前端'
    elif '软件' in ele:
        return '软件开发'
    elif '运维'in ele:
        return '运维'
    elif '数据库'in ele:
        return '数据库'
    elif '嵌入式'in ele:
        return '嵌入式'
    elif 'ios'in ele.lower():
        return 'IOS'
    elif( 'andriod'in ele.lower()) or ( '安卓'in ele.lower()):
        return 'andriod'
    

origin_data['岗位']=origin_data['岗位名称'].apply(arithmatic)
origin_data[origin_data['岗位'] == '数据库'].shape
```

```python
#统计语言
languages = ['java','c++','go','python','js','html','c#']
def arithmatic(ele):
    if 'java' in ele.lower():
        return 'java'
    elif ('c++' in ele.lower()):
        return 'c++'
    elif ('go' in ele.lower()):
        return 'go'
    elif ('python' in ele.lower()):
        return 'python'
    elif ('html' in ele.lower()):
        return 'html'
    elif ('html' in ele.lower()):
        return 'html'
    elif ('c#' in ele.lower()):
        return 'c#'
    elif ('js' in ele.lower()) or ('javascript' in ele.lower()):
        return 'js'

origin_data['语言']=origin_data['岗位名称'].apply(arithmatic)
origin_data.groupby('语言').agg({'城市':'count'})
```

```python
#统计最低，最高，平均薪资
def lowsalary(ele):
    return int(ele.split('-')[0].split('k')[0]+'000')
origin_data['最低工资'] = origin_data['薪资待遇'].apply(lowsalary)
def highsalary(ele):
    return int(ele.split('-')[1].split('k')[0]+'000')
origin_data['最高工资'] = origin_data['薪资待遇'].apply(highsalary)
origin_data
```

```python
def avgsalary(ele):
    return (ele.最低工资+ele.最高工资)/2 #apply取属性值使用.+属性值的方式
origin_data['平均工资'] = origin_data.apply(avgsalary,axis = 1)
origin_data
```

```python
#统计经验要求和学历
def experience(ele):
    return ele.split('/')[0].strip()
def scholar(ele):
    return ele.split('/')[1].strip()
origin_data['经验要求'] = origin_data['应聘条件'].apply(experience)
origin_data['学历要求'] = origin_data['应聘条件'].apply(scholar)
origin_data
```
